```python
import pandas as pd
import requests
```

```python
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36',
    'Accept':'application/json, text/plain, */*',
    'Accept-Language': 'en-US,en;q=0.9,vi;q=0.8',
    'Referer': 'https://tiki.vn/noi-luo-c-ga-inox-3-day-sunhouse-sh28lg-p155640558.html?itm_campaign=CTP_YPD_TKA_PLA_UNK_ALL_UNK_UNK_UNK_UNK_X.180809_Y.1763136_Z.3778628_CN.%5BPRODUCT%5D-NOI-CAC-LOAI&itm_medium=CPC&itm_source=tiki-ads&spid=25420919',
    'x-guest-token': 'BQh69X4ZebcLO7AjCnDqVIYilpGEvPsa',
    'Connection': 'keep-alive',
    'TE': 'Trailers',
}

params = {
    'platform': 'web',
    'spid': '25420919',
    'version': '3'
}
```

```python
df =  pd.read_csv('Products_id_all.csv')
p_ids = df.id.to_list()
print(len(p_ids))
```

```python
def parser_product(json):
    d = dict()
    d['id'] = json.get('id')
    d['sku'] = json.get('sku')
    d['short_description'] = json.get('short_description')
    d['price'] = json.get('price')
    d['list_price'] = json.get('list_price')
    d['price_usd'] = json.get('price_usd')
    d['discount'] = json.get('discount')
    d['discount_rate'] = json.get('discount_rate')
    d['review_count'] = json.get('review_count')
    d['order_count'] = json.get('order_count')
    d['inventory_status'] = json.get('inventory_status')
    d['is_visible'] = json.get('is_visible')
    stock_item = json.get('stock_item')
    if stock_item is not None:
        d['stock_item_qty'] = stock_item.get('qty')
        d['stock_item_max_sale_qty'] = stock_item.get('max_sale_qty')
    else:
        d['stock_item_qty'] = None
        d['stock_item_max_sale_qty'] = None
    d['product_name'] = json.get('meta_title')
    d['brand_id'] = json.get('brand').get('id')
    d['brand_name'] = json.get('brand').get('name')
    d
    return d
```

```python
result = []
for pid in p_ids:
    response = requests.get('https://tiki.vn/api/v2/products/{}'.format(pid), headers=headers, params=params)
    if response.status_code == 200:
        print('Crawl data {} success !'.format(pid))
        response.encoding = 'utf-8'  # set the encoding to 'utf-8'
        result.append(parser_product(response.json()))
df_product = pd.DataFrame(result)
df_product.to_csv('crawled_data_product_all.csv', index=False)
```
